# CRIANDO O PROGRAMA

```python
print('-' * 20)
print('TABUADA'.center(20))
print('-' * 20)

numero = input('Qual número deseja saber a tabuada: ')

if numero.isnumeric :
    numero = int(numero)
    for v in range(1, 11):
        print('{0} * {1} = {2}'.format(numero, v, v * numero))
else:
    print('Número inválido!')
    print('Fim do programa')
print('-' * 20)

input('Pausa ...')
```
